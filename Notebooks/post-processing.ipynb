# Post-Processing of ROH calls

```python
import numpy as np
import pandas as pd
import os        
import matplotlib.pyplot as plt
```

```python
folder = "" # The Folder to Analyze
```

```python
def load_data(folder="./Simulated/Example0/"):
        """Load and return genetic Map and Posterior"""
        
        ### Linkage Map
        map_path = folder + "map.csv"
        
        if os.path.exists(map_path):
            r_map = np.loadtxt(
                map_path, dtype="float", delimiter=",")
        else:
            # Eventually: Runtime Warning
            print("No Genetic Map found. Defaulting...")
            r_map = np.arange(np.shape(self.ob_stat)[1])

        ### Load Posterior
        post_path = folder + "posterior.csv"
        posterior = np.loadtxt(post_path, dtype="float", delimiter=",")
        
        assert(len(r_map) == np.shape(posterior)[1])  # Sanity Check
            
        print(f"Successfully loaded from {folder}")
        return r_map, posterior
```

```python
r_map, posterior = load_data("../Empirical/Sard100_0-10kROH/")
```

# Do the Post-Processing

```python
cutoff = 0.8 # Cutoff Probability for ROH State

roh_post = 1 - np.exp(posterior[0,:])  # Calculate the ROH Posterior
l_cutoff = 0.01 # Cutoff in Morgan

roh = roh_post > cutoff
frac_roh = np.mean(roh)
print(f"Fraction Markers in ROH: {frac_roh:.4f}")
```

```python
### Identify Stretches:
x1 = np.hstack([ [False], roh, [False] ]).astype("int")  # padding
d = np.diff(x1)
starts = np.where(d == 1)[0]
ends = np.where(d == -1)[0]
l = ends - starts


ends_map = r_map[ends]  # Transform to Map Lengths
starts_map = r_map[starts]
l_map = ends_map - starts_map 
```

```python
l_map > l_cutoff # Which markers to use (use Length Cutoff)

full_df = pd.DataFrame({'Start':starts,'End':ends, 
            'StartM':starts_map,'EndM':ends_map, 'length':l,'lengthM':l_map})

df = full_df[full_df["lengthM"]>l_cutoff]
```

# Do some Figures

```python
lw =10

plt.figure(figsize=(14,4))
plt.plot(roh_post, "ro", ms=1)
hl = plt.hlines(y=[1.1]*len(df), xmin=df["Start"], xmax=df["End"], colors="green", linewidth=lw)
plt.show()
```
